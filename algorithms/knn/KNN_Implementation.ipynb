```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
np.random.seed(42)
apple_sizes=np.random.normal(600,200,15)
apple_prices=np.random.normal(3,0.5,15)
lemon_sizes=np.random.normal(400,200,15)
lemon_prices=np.random.normal(2,0.5,15)
watermelon_sizes=np.random.normal(300,200,15)
watermelon_prices=np.random.normal(1,0.5,15)

X=np.vstack(
    [np.column_stack((apple_sizes,apple_prices)),
    np.column_stack((lemon_sizes,lemon_prices)),
    np.column_stack((watermelon_sizes,watermelon_prices))]
)
#print(X)
y=np.array(["apple"]*15+["lemon"]*15+["watermelon"]*15)
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
new_fruit=np.array([[400,2]])
prediction=knn.predict(new_fruit)
print(prediction)

plt.figure(figsize=(12,6))
plt.scatter(apple_sizes,apple_prices,color="red",label="apple",s=90)
plt.scatter(lemon_sizes,lemon_prices,color="green",label="lemon",s=190)
plt.scatter(watermelon_sizes,watermelon_prices,color="blue",label="watermelon",s=200)
plt.scatter(new_fruit[0,0],new_fruit[0,1],color="black",marker="*",s=1000)
plt.xlabel("size")
plt.ylabel("price")
plt.legend()
plt.grid(True)
plt.show()
```
