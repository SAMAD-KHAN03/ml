```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
data={
    "age":[25,30,35,40,45,50,55,60],
    "salary":[50000,60000,70000,80000,90000,100000,110000,120000],
    "prev_loan":[1,0,1,0,1,0,1,0],
    "cbil_score":[700,750,800,850,900,950,1000,1050],
    "loan_approved":[1,1,0,1,0,1,0,1]
}
df=pd.DataFrame(data)
print(df)
x = df[["age","salary","prev_loan","cbil_score"]]
y = df["loan_approved"]
model=DecisionTreeClassifier(max_depth=3)
model.fit(x,y)
new_entry = [[30,45000,1,780]]
#model.predict(new_entry)
print(model.predict(new_entry))
plt.figure(figsize=(12,6))
plot_tree(
    model,
    feature_names=x.columns,
    class_names=["Rejected","Approved"],
    filled=True
)
plt.show()
```
