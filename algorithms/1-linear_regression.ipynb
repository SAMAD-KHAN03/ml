```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
import joblib
from sklearn.metrics import mean_squared_error
df=pd.read_csv("scattered_house_price_data.csv")
area=df['flat_area']
price=df['price_inr']
plt.figure(figsize=[10,6])
plt.scatter(area,price,color="blue",marker=".",label="data points")

# here w will train with linear regression
model=LinearRegression()
model.fit(area.values.reshape(-1,1),price)#yhi line training h sara math sab kuch iske andar ho jaega
# arr = np.array([1, 2, 3, 4, 5, 6])
# reshaped_arr = arr.reshape(-1, 1)
# print(reshaped_arr)
# Output:
# [[1]
#  [2]
#  [3]
#  [4]
#  [5]
#  [6]]
prediction=model.predict(area.values.reshape(-1,1))
mse=mean_squared_error(prediction,price) 
print(mse)
plt.plot(area,prediction,color="red",linewidth=2)#area elements will be x values and 
plt.legend()


plt.grid(True)
plt.show()
joblib.dump(model,"bangalore-housing.pkl")
# The ".pkl" file extension typically indicates a Python pickle file. These files are created using Python's pickle module, which serializes Python objects into a byte stream. This process is known as "pickling." The inverse operation, converting the byte stream back into a Python object, is called "unpickling."
```
