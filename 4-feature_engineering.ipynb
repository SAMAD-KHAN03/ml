```python
! pip install -U scikit-learn
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

data = {
    'Color': ['red', 'green', 'blue', 'green', 'red'],
    'Size': ['S', 'M', 'L', 'M', 'S'],
    # 'InStock': ['yes', 'no', 'yes', 'no', 'yes']
}

df=pd.DataFrame(data)
# df=LabelEncoder
# print(df)
 
# one hot encoding 
# Color = ['red', 'green', 'blue', 'green', 'red']
# after one hot endcoding
#    red  green  blue
# 0    1      0     0
# 1    0      1     0
# 2    0      0     1
# 3    0      1     0
# 4    1      0     0
	# Each row has a 1 in the column of its category, and 0 elsewhere.
	# â€¢	No implied order between categories (unlike label encoding).

# encoding
ohe=OneHotEncoder(sparse_output=False)
color_encoded=ohe.fit_transform(df[['Color']])#applying one hot encoding on color column of dataframe
color_df=pd.DataFrame(color_encoded,columns=ohe.get_feature_names_out(['Color']))

size_order=[['S','M','L']]#iss order ke hisab se 0,1,2,3 etc assign hoga
ord_encoding=OrdinalEncoder(categories=size_order)
size_encoded=ord_encoding.fit_transform(df[['Size']])
size_df=pd.DataFrame(size_encoded,columns=["Size_encoded"])

final_df=pd.concat([df,color_df,size_df],axis=1)
print(final_df)


```
