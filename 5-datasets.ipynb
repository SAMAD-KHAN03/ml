```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
df=pd.DataFrame(housing.data,columns=housing.feature_names)
df["MedianHouseValue"]=housing.target
# In Scikit-learn, when working with datasets, the .target property refers to the dependent variable or the response variable that a machine learning model aims to predict.


# print(df)
df.head(20000)
```

```python
import matplotlib.pyplot as plt
x=df.HouseAge.head(10)
y=df.MedInc.head(10)
plt.plot(x,y)
plt.title("House Age vs MedInc graph")
plt.xlabel("HouseAge")
plt.ylabel("Medinc")
plt.show()
df.hist(bins=50,figsize=[20,15])
plt.show()
```
